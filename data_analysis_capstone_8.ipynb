```python
likert_cols = ['membership_value', 'digital_value', 'participation', 'family_recommendation', 'recommendation_based_trial', 'arina_digital_experience', 'shopping_habit', 'arina_tech_support', 'digital_use', 'online_offline_experience', 'digital_effort']
numerical_cols = ['age']
categorical_cols = ['language', 'Education', 'gender', 'speak_finnish']
```

```python
# Verify that all likert_cols are in the DataFrame columns
missing_cols = [col for col in likert_cols if col not in df.columns]
if missing_cols:
    print(f"These columns are missing from the DataFrame: {missing_cols}")
else:
    # Fill missing values for Likert scale columns with 1
    for col in likert_cols:
        df[col] = df[col].fillna(1)
```

```python
# Identify and replace non-numeric values in numerical columns with NaN
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Fill NaN values in numerical columns with the median
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# For other columns, fill missing values in categorical columns with 'Unknown'
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
# Check the dataframe shape after handling missing values
print(df.shape)
```

```python
# Handle specific columns with missing values
# Fill missing values in 'timestamp' with a default value or drop these rows
df['timestamp'] = df['timestamp'].fillna('2022-01-01 00:00:00')  # Example default value, adjust as needed

# Fill missing values in 'motivation_text' with 'No response'
df['motivation_text'] = df['motivation_text'].fillna('No response')

# Fill missing values in 'additional_comment' with 'No comment'
df['additional_comment'] = df['additional_comment'].fillna('No comment')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Map gender values for clarity (1=Female, 2=Male, 3=Other)
gender_map = {1: 'Female', 2: 'Male', 3: 'Other'}
df['gender'] = df['gender'].map(gender_map)

# Ensure Likert columns are numeric
for col in likert_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Reset index to avoid duplicate labels if needed
df = df.reset_index(drop=True)

# Set up matplotlib figure size
plt.figure(figsize=(15, 10))

# Histogram for age distribution with percentages
plt.subplot(2, 2, 1)
ax = sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution')
for p in ax.patches:
    height = p.get_height()
    total = len(df['age'].dropna())
    percent = f'{(height / total) * 100:.1f}%'
    ax.text(p.get_x() + p.get_width() / 2., height, percent, ha="center")

# Bar plot for gender distribution with percentages (Including 'Other')
plt.subplot(2, 2, 2)
ax = sns.countplot(x='gender', data=df, palette='Set2', order=['Female', 'Male', 'Other'])  # Include 'Other'
plt.title('Gender Distribution')
total = len(df['gender'].dropna())
for p in ax.patches:
    height = p.get_height()
    percent = f'{(height / total) * 100:.1f}%'
    ax.text(p.get_x() + p.get_width() / 2., height, percent, ha="center")

# Bar plot for language distribution with percentages (Clean and colorful)
plt.subplot(2, 2, 3)
ax = sns.countplot(x='language', data=df, palette='muted')  # 'muted' palette for cleaner look
plt.title('Language Distribution')

# Rotate x-axis labels if they are long
plt.xticks(rotation=45, ha="right")

# Add percentages on bars
total = len(df['language'].dropna())
for p in ax.patches:
    height = p.get_height()
    percent = f'{(height / total) * 100:.1f}%'
    ax.text(p.get_x() + p.get_width() / 2., height, percent, ha="center")

# Calculate correlation matrix without FutureWarning
corr_matrix = df[likert_cols].select_dtypes(include='number').corr()

# Heatmap for correlations among Likert scale columns
plt.subplot(2, 2, 4)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Likert Scale Columns')

plt.tight_layout()
plt.show()

# Plot the frequency counts for each Likert column with percentages
for column in likert_cols:
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(
        x=column,
        data=df,
        palette='viridis',
        order=sorted(df[column].dropna().unique())
    )
    plt.title(f'Distribution of Responses for {column}')
    plt.xlabel('Likert Scale')
    plt.ylabel('Frequency')

    # Set Likert scale labels directly with correct tick positions
    tick_positions = [0, 1, 2, 3, 4]  # Positions of Likert scale responses (1 to 5)
    tick_labels = ['Strongly disagree', 'Somewhat disagree', 'Neutral', 'Somewhat agree', 'Strongly agree']
    plt.xticks(ticks=tick_positions, labels=tick_labels)

    # Add percentages on bars
    total = len(df[column].dropna())
    for p in ax.patches:
        height = p.get_height()
        percent = f'{(height / total) * 100:.1f}%'
        ax.text(p.get_x() + p.get_width() / 2., height, percent, ha="center")

    plt.show()
```

```python
# Advanced Analysis - Factor Analysis
```

```python
from sklearn.decomposition import FactorAnalysis
from sklearn.impute import SimpleImputer

# Ensure Likert columns are numeric
for col in likert_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Impute missing values in Likert scale columns with the median
imputer = SimpleImputer(strategy='median')
df[likert_cols] = imputer.fit_transform(df[likert_cols])

# Perform factor analysis
fa = FactorAnalysis(n_components=3)
likert_factors = fa.fit_transform(df[likert_cols])

# Add factors to the dataframe
df['Factor1'] = likert_factors[:, 0]
df['Factor2'] = likert_factors[:, 1]
df['Factor3'] = likert_factors[:, 2]

# Print the factor loadings
factor_loadings = pd.DataFrame(fa.components_, columns=likert_cols)
print(factor_loadings)
```

```python
# Cluster Analysis - Perform clustering to identify member segments
```

```python
from sklearn.cluster import KMeans

# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[likert_cols])

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Factor1', y='Factor2', hue='Cluster', data=df, palette='viridis')
plt.title('Clusters of Members Based on Likert Scale Responses')
plt.show()
```
